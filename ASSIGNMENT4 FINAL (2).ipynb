```python
# Function to get distribution of specified column
def distribution(df_f, group_column_name):
    dist_df = df_f["channeltype"].value_counts()
    return dist_df
```

```python
#Calling function distribution by passing first 1000 rows dataframe and column name
df_dist = distribution(df_first_1000_rows, "channeltype")
```
